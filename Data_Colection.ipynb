```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
haar_data = cv2.CascadeClassifier('spec.xml')
```

# Live Face Data Collection Without SPECS 

```python
capture = cv2.VideoCapture(0)
Withoutspecs_data = []
while True :
    flag, img = capture.read()
    if flag :
        faces = haar_data.detectMultiScale(img)
        for x,y,w,h in faces : 
            cv2.rectangle(img, (x,y), (x+w, y+h), (245,224,69), 2)
            face = img[y: y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(Withoutspecs_data))
            if len(Withoutspecs_data) < 200 :
                Withoutspecs_data.append(face)
        cv2.imshow("Capturing Without Specs Images ..!",img)
        if cv2.waitKey(2) == 27 or len(Withoutspecs_data) >= 200 :
            break

capture.release()
cv2.destroyAllWindows()
```

```python
np.save("Without_specs.npy",Withoutspecs_data)
```

```python
np.save("With_specs.npy",Withspecs_data)
```
